<b> Feel free to any method to complete this tasks. However, we recommend you use either list filtering [], or .loc to complete this task.</b>

```python
df_filter = dataframe[[' State-gov', ' Bachelors', ' Never-married', ' Adm-clerical',
                      ' Not-in-family', ' White', ' Male', ' United-States', ' <=50K']]

print(df_filter)
```

<b> Q2. Please re-name all the newly filtered columns in the pandas DataFrame to the following: </b>

Employment Type, Degree Status, Marriage-Status, Job-Role, Family-Role, Ethnicity, Gender, Country, Earnings

E.g. State-Gov becomes Employment Type, Bachelors becomes Degree Status, etc.

<b> Put your code below </b>

```python
df_filter.rename(columns={' State-gov': 'Employment Type', ' Bachelors': 'Degree Status',
                         ' Never-married': 'Marriage-Status', ' Adm-clerical': 'Job-Role',
                         ' Not-in-family': 'Family-Role', ' White': 'Ethnicity',
                         ' Male': 'Gender', ' United-States': 'Country', ' <=50K': 'Earnings'}, inplace=True)

print(df_filter)
```

<b> Q3. The Job Role Columns holds the job information for each individual in this census snapshot. Using this column, create a Bar Chart that shows the count of 'Unique' Jobs per Job Group in the "Job-Role" Column in ascending order, as per the provided image below </b>


<b> Put your code below </b>

```python
unique_jobs = df_filter['Job-Role'].value_counts().sort_values()
jobs = df_filter['Job-Role'].unique()

ax = unique_jobs.plot(kind='bar', xlabel='Job-Role', ylabel='Total # of People')

plt.show()
```

<b> Q4. Please create two bar plots as per below that show:
    
    1) The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
    2) The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States 

Please note you will be looking specifically at the *Job Role* column

<b> Put Your Code Below </b>

```python
hsg = df_filter[df_filter['Degree Status']==' HS-grad']
hsg_50k_down = hsg[hsg['Earnings']==' <=50K']
hsg_50k_down_jobs = hsg_50k_down['Job-Role'].value_counts()

ax = hsg_50k_down_jobs.plot(kind='bar',xlabel='Job-Role',ylabel='Total # of People')

plt.show()
```

```python
hsg_50k_up = hsg[hsg['Earnings']==' >50K']
hsg_50k_up_jobs = hsg_50k_up['Job-Role'].value_counts()

ax = hsg_50k_up_jobs.plot(kind='bar',xlabel='Job-Role',ylabel='Total # of People')

plt.show()
```

<H2> Challenge Question </H2>

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
e50k_up = df_filter[df_filter['Earnings']==' >50K']
jobs_50k_up = e50k_up['Job-Role'].value_counts()

jobs = df_filter['Job-Role'].value_counts()

proportions_50k_up = jobs_50k_up/jobs

ax = proportions_50k_up.sort_values(ascending=False).plot(kind='bar',xlabel='Job-Role',ylabel='% of Jobs over 50K')

plt.show()
```
